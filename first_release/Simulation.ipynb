```python
# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                       [ INFORMAÇÕES ]                                     |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# ============================================================================================= 
#
# =============================================================================================
# |                                                                                           |
# |                                          [ AUTOR ]                                        |
# |                                                                                           |
# ============================================================================================= 
# Nome: RAFAEL ZACARIAS PALIERINI
# Curso: CIÊNCIA DA COMPUTAÇÃO
# R.A: 22.218.030-9
# 
#
# =============================================================================================
# |                                                                                           |
# |                              [ PROJETO DE INICIAÇÃO DIDÁTICA ]                            |
# |                                                                                           |
# ============================================================================================= 
# Nome do Projeto: ELABORAÇÃO DE SIMULAÇÕES PARA LABORATÓRIOS DE FÍSICA 1
# Orientadora: ELIANE DE FÁTIMA CHINAGLIA
# Universidade: CENTRO UNIVERSITÁRIO FEI
# Cidade: SÃO BERNARDO DO CAMPO
#
# =============================================================================================
# |                                                                                           |
# |                                          [ DATA ]                                         |
# |                                                                                           |
# ============================================================================================= 
# Data de início: 01 de Setembro de 2020
# Ultíma atualização: 23 de Janeiro de 2021











# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                         [ SUMÁRIO ]                                       |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
#
#   # - NOME...................................................................................
# =============================================================================================
#
#   1 - IMPORTS................................................................................
#       1.1 - GUI..............................................................................
#       1.2 - FUNÇÕES..........................................................................
#   2 - CLASSES................................................................................
#       2.1 - BOTÕES...........................................................................
#       2.2 - CONTROLADOR DO GRÁFICO...........................................................
#       2.3 - GRÁFICO..........................................................................
#       2.4 - DIVISÃO..........................................................................
#   3 - FUNÇÕES................................................................................
#       3.1 - NOVO FRAME.......................................................................
#       3.2 - MOSTRA GRÁFICO...................................................................
#       3.3 - ESCONDE GRÁFICO..................................................................
#       3.4 - CALCULA TRAJETÓRIA...............................................................
#       3.5 - CALCULA POSIÇÃO x(t).............................................................
#       3.6 - CALCULA POSIÇÃO y(t).............................................................
#       3.7 - CALCULA VELOCIDADE x(t)..........................................................
#       3.8 - CALCULA VELOCIDADE y(t)..........................................................
#       3.9 - CALCULA ACELERAÇÃO x(t)..........................................................
#       3.10 - CALCULA ACELERAÇÃO y(t).........................................................
#       3.11 - REMOVE UM PLOT..................................................................
#       3.12 - LIMPA O GRÁFICO.................................................................
#       3.13 - IMPORTA VALORES EXPERIMENTAIS...................................................
#   4 - MAIN...................................................................................
#       4.1 - ROOT.............................................................................
#           4.1.1 - INICIALIZA AS DIVISÕES.....................................................
#           4.1.2 - FUNÇÕES PARA BOTÕES........................................................
#           4.1.3 - BOTÕES DO MENU.............................................................
#           4.1.4 - TRAJETÓRIA.................................................................
#           4.1.5 - POSIÇÃO x(t)...............................................................
#           4.1.6 - POSIÇÃO y(t)...............................................................
#           4.1.7 - VELOCIDADE x(t)............................................................
#           4.1.8 - VELOCIDADE y(t)............................................................
#           4.1.9 - CALCULA ACELERAÇÃO x(t)....................................................
#           4.1.10 - CALCULA ACELERAÇÃO y(t)...................................................
#           4.1.11 - GRÁFICOS DAS DIVISÕES.....................................................
#           4.1.12 - ATUALIZA AS DIVISÕES......................................................
#           4.1.13 - BIND DOS BOTÕES...........................................................











# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                         [ IMPORTS ]                                       |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# Aqui é o local aonde deve ficar todos os imports necessários para a Interface Gráfica
# de Usuário (GUI) funcionar de maneira adequada.
# 

import os
import sys
import webbrowser
sys.path.insert(1, os.getcwd()+"\\CustomLibraries")

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                          || GUI ||                                        |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
from tkinter import *
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, NavigationToolbar2Tk)
from matplotlib.backend_bases import key_press_handler
from matplotlib.figure import Figure

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                        || FUNÇÕES ||                                      |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
from equacoes import *
import csvie
import time
import numpy as np































# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                         [ CLASSES ]                                       |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
#
# Aqui ficam as classes do código, cada classe tem uma explicação em forma de comentário
# em cima da própria classe. 
#

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                        || BOTÕES ||                                       |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# Está classe é responsável por armazenar todas as informações importantes do botão.
class AlternativeButton:
    def __init__(self, x, y, width, height, background, function):
        self.x = x
        self.y = y
        self.width = width
        self.height = height
        self.idle = PhotoImage(file=os.getcwd()+"\\Images\\Menu_Buttons\\" + background + ".png")
        self.clicked = PhotoImage(file=os.getcwd()+"\\Images\\Menu_Buttons\\" + background + "_Hold.png")
        self.over = PhotoImage(file=os.getcwd()+"\\Images\\Menu_Buttons\\" + background + "_Over.png")
        self.using = self.idle
        self.function = function
        return
    
    def isIdle(self):
        self.using = self.idle
        return

    def isClicked(self):
        self.using = self.clicked
        return

    def isOver(self):
        self.using = self.over
        return

    def run(self):
        return self.function()

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                || CONTROLADOR DO GRÁFICO ||                               |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# Está classe tem como objetivo fornecer controles ao usuário sobre o gráfico. Como por
# exemplo, movimentar, aproximar ou afastar o gráfico.
# 
#
#
# Créditos ao 'seadoodude', usuário do StackOverflow
#
# Link: https://stackoverflow.com/questions/11551049/matplotlib-plot-zooming-with-scroll-wheel
#
# Data de Acesso: 10:47 AM   29/03/2020
class GraphController:
    def __init__(self):
        self.press = None
        self.cur_xlim = None
        self.cur_ylim = None
        self.x0 = None
        self.y0 = None
        self.x1 = None
        self.y1 = None
        self.xpress = None
        self.ypress = None
        
        return
    
    def __del__(self):
        print("Controller destroyed!")
    
    # (ZOOM)
    def zoom_factory(self, graph, base_scale = 2):
        def zoom(event):
            cur_xlim = graph.get_xlim()
            cur_ylim = graph.get_ylim()

            xdata = event.xdata # get event x location
            ydata = event.ydata # get event y location

            if event.button == 'up':
                # deal with zoom in
                scale_factor = 1 / base_scale
            elif event.button == 'down':
                # deal with zoom out
                scale_factor = base_scale
            else:
                # deal with something that should never happen
                scale_factor = 1
                print (event.button)

            new_width = (cur_xlim[1] - cur_xlim[0]) * scale_factor
            new_height = (cur_ylim[1] - cur_ylim[0]) * scale_factor

            relx = (cur_xlim[1] - xdata)/(cur_xlim[1] - cur_xlim[0])
            rely = (cur_ylim[1] - ydata)/(cur_ylim[1] - cur_ylim[0])

            graph.set_xlim([xdata - new_width * (1-relx), xdata + new_width * (relx)])
            graph.set_ylim([ydata - new_height * (1-rely), ydata + new_height * (rely)])
            graph.figure.canvas.draw()

        fig = graph.get_figure() # get the figure of interest
        fig.canvas.mpl_connect('scroll_event', zoom)

        return zoom
    
    # (MOVIMENTAÇÃO)
    def pan_factory(self, graph):
        def onPress(event):
            if event.inaxes != graph: return
            self.cur_xlim = graph.get_xlim()
            self.cur_ylim = graph.get_ylim()
            self.press = self.x0, self.y0, event.xdata, event.ydata
            self.x0, self.y0, self.xpress, self.ypress = self.press

        def onRelease(event):
            self.press = None
            graph.figure.canvas.draw()

        def onMotion(event):
            if self.press is None: return
            if event.inaxes != graph: return
            dx = event.xdata - self.xpress
            dy = event.ydata - self.ypress
            self.cur_xlim -= dx
            self.cur_ylim -= dy
            graph.set_xlim(self.cur_xlim)
            graph.set_ylim(self.cur_ylim)

            graph.figure.canvas.draw()

        fig = graph.get_figure() # get the figure of interest

        # attach the call back
        fig.canvas.mpl_connect('button_press_event',onPress)
        fig.canvas.mpl_connect('button_release_event',onRelease)
        fig.canvas.mpl_connect('motion_notify_event',onMotion)

        #return the function
        return onMotion

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                       || GRÁFICO ||                                       |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# Está classe é responsável por criar um gráfico no frame desejado.
class Graph:
    def __init__(self, name, xLabel, yLabel, frame):
        self.graphName = name
        self.xLabel = xLabel
        self.yLabel = yLabel
        self.plotName = []
        self.xPoint = []
        self.yPoint = []
        self.typeGraph = []

        # CRIANDO O GRÁFICO
        # 
        # Aqui é definido as configurações principais do nosso gráfico, como a quantidade
        # de DPI, tamanho do gráfico, o seu subplot e a cor de fora do gráfico.
        self.fig = Figure(figsize=(8, 4.4), dpi=82, facecolor="#1A4364")
        self.graph = self.fig.add_subplot(111)

        # ESTILO DO GRÁFICO
        #
        # Essa parte é responsável pelo design do Gráfico, como cores, tipo de grid, nome
        # do gráfico e o nome dos eixos.
        self.graph.grid(color = "#dbdbdb")
        self.graph.tick_params(axis = "both", color = "white", labelcolor = "white")
        self.graph.spines['bottom'].set_color("white")
        self.graph.spines['top'].set_color("white")
        self.graph.spines['right'].set_color("white")
        self.graph.spines['left'].set_color("white") 
        self.graph.set_ylabel(yLabel, color = "white")
        self.graph.set_xlabel(xLabel, color = "white")
        self.graph.set_title(self.graphName, color = "white", fontweight='bold')

        # CANVAS DO GRÁFICO
        # 
        # O canvas do gráfico é responsável por permitir que seja possível visualizar o
        # gráfico do matplotlib em um GUI com Tkinter.
        self.graphCanvas = FigureCanvasTkAgg(self.fig, master=frame)
        self.graphCanvas.draw()
        
        # CONTROLADOR DO GRÁFICO
        # 
        # Essa parte é responsável por permitir que o gráfico, introduzido na Interface
        # Gráfica de Usuário (GUI), seja controlado pelo controlador da classe já cria-
        # da e nomeada de GraphController.
        self.graphScale = 1.1
        self.controller = GraphController()
        self.figZoom = self.controller.zoom_factory(self.graph, base_scale = self.graphScale)
        self.figPan = self.controller.pan_factory(self.graph)

        # POSIÇÃO DO GRÁFICO
        # 
        # Essa linha é responsável por posicionar aonde ficará o nosso gráfico.
        self.graphCanvas.get_tk_widget().pack(side=BOTTOM, fill=BOTH, expand=1)
        
        return

    def __del__(self):
        del self.graphCanvas
        del self.controller
        del self.graph
        del self.fig
        return
    
    def plot(self, x, y, plotName, typeGraph = "Normal", animated = "Yes"):

        #Váriavel de controle para a velocidade da animação de plot
        iterator = 0.05*len(x)

        if(animated == "Yes"):
            for i in range(len(x)):
                if iterator <= 0:
                    iterator = 0.05*len(x)

                    self.graph.clear()
                    self.graph.grid(color = "#dbdbdb")
                    self.graph.tick_params(axis = "both", color = "white", labelcolor = "white")
                    self.graph.spines['bottom'].set_color("white")
                    self.graph.spines['top'].set_color("white")
                    self.graph.spines['right'].set_color("white")
                    self.graph.spines['left'].set_color("white") 
                    self.graph.set_ylabel(self.yLabel, color = "white")
                    self.graph.set_xlabel(self.xLabel, color = "white")
                    self.graph.set_title(self.graphName, color = "white", fontweight='bold')

                    subX = []
                    subY = []

                    for j in range(i):
                        subX.append(x[j])
                        subY.append(y[j])

                    for j in range(len(self.xPoint)):
                        if(self.typeGraph[j] == "Experimental"):
                            self.graph.plot(self.xPoint[j], self.yPoint[j], 'o', label=self.plotName[j])
                        else:
                            self.graph.plot(self.xPoint[j], self.yPoint[j], label=self.plotName[j])

                    if(typeGraph == "Experimental"):
                        self.graph.plot(subX, subY, 'o', label= plotName)
                    else:
                        self.graph.plot(subX, subY, label= plotName)
                    self.graph.legend(loc="upper right")
                    self.graphCanvas.draw()

                else:
                    iterator -= 1

        self.graph.clear()
        self.graph.grid(color = "#dbdbdb")
        self.graph.tick_params(axis = "both", color = "white", labelcolor = "white")
        self.graph.spines['bottom'].set_color("white")
        self.graph.spines['top'].set_color("white")
        self.graph.spines['right'].set_color("white")
        self.graph.spines['left'].set_color("white") 
        self.graph.set_ylabel(self.yLabel, color = "white")
        self.graph.set_xlabel(self.xLabel, color = "white")
        self.graph.set_title(self.graphName, color = "white", fontweight='bold')

        for i in range(len(self.xPoint)):
            if(self.typeGraph[i] == "Experimental"):
                self.graph.plot(self.xPoint[i], self.yPoint[i], 'o', label=self.plotName[i])
            else:
                self.graph.plot(self.xPoint[i], self.yPoint[i], label=self.plotName[i])

        if (typeGraph == "Experimental"):
            self.graph.plot(x, y, 'o', label=plotName)
        else:
            self.graph.plot(x, y, label=plotName)
        
        self.graph.legend(loc="upper right")
        self.graphCanvas.draw()
        self.xPoint.append(x)
        self.yPoint.append(y)
        self.plotName.append(plotName)
        self.typeGraph.append(typeGraph)

    def delPlot(self, plotName):
        index = self.plotName.index(plotName)
        self.plotName.pop(index)
        self.xPoint.pop(index)
        self.yPoint.pop(index)
        self.typeGraph.pop(index)

        self.graph.clear()
        self.graph.grid(color = "#dbdbdb")
        self.graph.tick_params(axis = "both", color = "white", labelcolor = "white")
        self.graph.spines['bottom'].set_color("white")
        self.graph.spines['top'].set_color("white")
        self.graph.spines['right'].set_color("white")
        self.graph.spines['left'].set_color("white") 
        self.graph.set_ylabel(self.yLabel, color = "white")
        self.graph.set_xlabel(self.xLabel, color = "white")
        self.graph.set_title(self.graphName, color = "white", fontweight='bold')

        for j in range(len(self.xPoint)):
            if(self.typeGraph[j] == "Experimental"):
                self.graph.plot(self.xPoint[j], self.yPoint[j], 'o', label=self.plotName[j])
            else:
                self.graph.plot(self.xPoint[j], self.yPoint[j], label=self.plotName[j])
        
        if self.xPoint != []:
            self.graph.legend(loc="upper right")
        self.graphCanvas.draw()

    def clearPlot(self):
        self.plotName = []
        self.xPoint = []
        self.yPoint = []
        self.typeGraph = []

        self.graph.clear()
        self.graph.grid(color = "#dbdbdb")
        self.graph.tick_params(axis = "both", color = "white", labelcolor = "white")
        self.graph.spines['bottom'].set_color("white")
        self.graph.spines['top'].set_color("white")
        self.graph.spines['right'].set_color("white")
        self.graph.spines['left'].set_color("white") 
        self.graph.set_ylabel(self.yLabel, color = "white")
        self.graph.set_xlabel(self.xLabel, color = "white")
        self.graph.set_title(self.graphName, color = "white", fontweight='bold')
        self.graphCanvas.draw()

        

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                       || DIVISÃO ||                                       |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
#
# Como o software possuí muitas funcionalidades, é ideal que elas sejam separadas em di-
# visões. Desta forma, podemos implementar e modificar recursos sem prejudicar as funcio-
# nalidades já presentes no software.

class Division:
    # CONSTRUTOR DA DIVISÃO
    # 
    # O construtor é responsável por criar todos os componentes principais de uma divisão.
    def __init__(self, root, divWidth, divHeight, x, y, background):
        if(background.find("#") != -1):
            self.frame = Frame(root, width=divWidth, height=divHeight, bg=background)
        else:
            self.image = PhotoImage(file=os.getcwd()+"\\Images\\Background\\" + background)
            self.frame = Frame(root, width=divWidth, height=divHeight)
            self.background = Canvas(self.frame, width=divWidth, height=divHeight, bd=0, highlightthickness=0)
            self.background.create_image(0, 0, image = self.image, anchor=NW)
            self.background.place(x=0, y=0, anchor=NW)

        self.frame.place(x=x, y=y, anchor=NW)

        self.xPlace = x
        self.yPlace = y
        self.divWidth = divWidth
        self.divHeight = divHeight
        self.labels = []
        self.entrys = []
        self.inputs = []
        self.buttons = []
        self.graph = 0

        root.update()

        #self.killButton = Button(self.frame, text="Apagar", command=self.kill)
        #self.killButton.place(x=200, y=100, anchor=CENTER)

        self.frame.update()
        return

    # DESTRUTOR DA DIVISÃO
    # 
    # O destrutor é responsável por apagar a divisão por completo, garantindo que seja li-
    # berado espaço na memória.
    def __del__(self):
        if(self.graph != 0):
            del self.graph
        return
    
    # CALLBACK DO DESTRUTOR
    # 
    # Chama o destrutor da própria classe, garantindo a destruição da mesma.
    def kill(self):
        return self.__del__()

    # ATUALIZA A DIVISÃO:
    # 
    # Está função é fundamental para o funcionamento da classe. Pois, ela é a responsável
    # por atualizar toda a divisão caso ocorra alguma alteração (novo botão, entry ou la-
    # bel) nela.
    def updateDivision(self):
        self.background.delete("all")
        if(self.image):
            self.background.create_image(0, 0, image = self.image, anchor=NW)
        self.drawButton()
        self.background.update()
        return

    # CRIA UMA NOVA LABEL
    # 
    # Está função é responsável por criar novas Labels dentro da divisão.
    def newLabel(self, name, width, x, y, background, link):
        if(name == "MetodoNum"):
            self.entrys.append([name, Label(self.frame,
                                            text="",
                                            bg="#1A4364",
                                            fg="#FFFFFF",
                                            font=("Helvetica", 8, "bold"),  
                                            bd=0)])

            self.entrys[-1][1].config(width = width, height = 1, text="Método: Numérico")

            self.entrys[-1][1].place(x=x, y=y, anchor=NW)

            return

        if(name == "MetodoAna"):
            self.entrys.append([name, Label(self.frame,
                                            text="",
                                            bg="#1A4364",
                                            fg="#FFFFFF",
                                            font=("Helvetica", 8, "bold"),  
                                            bd=0)])

            self.entrys[-1][1].config(width = width, height = 1, text="Método: Analítico")

            self.entrys[-1][1].place(x=x, y=y, anchor=NW)

            return
        
        elif(background.find("#") == -1):
            img = PhotoImage(file=os.getcwd() + "\\Images\\Label\\" + background)
            self.labels.append([name, Label(self.frame, image=img, bd=0)])
            self.labels[-1][1].photo = img

        else:
            self.labels.append([name, Label(self.frame, text=name, bg=background, bd=0)])

        self.labels[-1][1].place(x=x, y=y, anchor=NW)

        self.frame.update()

        if(name == "Resultado"):
            self.entrys.append([name, Label(self.frame,
                                            text="",
                                            bg="#12314A",
                                            fg="#FFFFFF",
                                            font=("Helvetica", 16),  
                                            bd=0)])

            self.entrys[-1][1].config(width = width, height = 1)

            newX= (5 + x + int(self.labels[-1][1].winfo_width()))
            self.entrys[-1][1].place(x=newX, y=y, anchor=NW)

        elif(link == True):
            self.newEntry(name, width, x, y, background, True)
            self.inputs.append([self.labels[-1], self.entrys[-1]])
        
        return

    # NOVAS ENTRYS
    # 
    # Está função é responsável por criar novas Entrys dentro da divisão.
    def newEntry(self, name, width, x, y, background, link):
        if(name == "Nome"):
            self.entrys.append([name]) 
            self.entrys[-1].append(Entry(self.frame, 
                                     bd=0, 
                                     bg="#1A4364", 
                                     fg="#FFFFFF", 
                                     width=width, 
                                     font=("Helvetica", 13), 
                                     justify=LEFT, 
                                     insertbackground="white"))
        
        else:
            self.entrys.append([name])
            self.entrys[-1].append(Entry(self.frame, 
                                    bd=0, 
                                    bg="#1A4364", 
                                    fg="#FFFFFF", 
                                    width=width, 
                                    font=("Helvetica", 16), 
                                    justify=RIGHT,
                                    insertbackground="white"))


        if (name == "Limite" or name == "Nome"):
            newX= (10 + x + int(self.labels[-1][1].winfo_width()))
        
        elif(link==True):
            newX= (5 + x + int(self.labels[-1][1].winfo_width()))

        self.entrys[-1][1].place(x=newX, y=y, anchor=NW)

        self.frame.update()

        return

    # CRIA UM BOTÃO NA DIVISÃO
    # 
    # Essa função é responsável por adicionar um novo botão com a propriedades que foram
    # passadas por parâmetro e adiciona-lo na lista de botões.
    def newButton(self, x, y, width, height, background, function):
        self.buttons.append(AlternativeButton(x, y, width, height, background, function))
        return
    
    # DESENHA BOTÃO
    # 
    # Essa função é responsável por desenhar todos os botões da divisão.
    def drawButton(self):
        for i in range(len(self.buttons)):
                self.background.create_image(self.buttons[i].x,
                                            self.buttons[i].y,
                                            image = self.buttons[i].using,
                                            anchor=NW)
        return
    
    # ANALISA SE O MOUSE ESTÁ SOBRE UM BOTÃO
    # 
    # Está função verifica a posição do mouse na tela e apartir disso ele verifica se hou-
    # ve a invasão da área do botão para então mudar a sua imagem.
    def isOver(self, event):
        time.sleep(0.035)
        for i in range(len(self.buttons)):
            if((event.x >= (self.buttons[i].x + 2)) 
                and (event.x <= (self.buttons[i].width + self.buttons[i].x - 2))
                and (event.y >= (self.buttons[i].y + 2)) 
                and (event.y <= (self.buttons[i].height + self.buttons[i].y - 2))):
                self.buttons[i].isOver()

            else:
                self.buttons[i].isIdle()
        
        self.updateDivision()

        return

    # EXECUTA A FUNÇÃO DO BOTÃO
    # 
    # Está função verifica a posição do mouse na tela e apartir disso ele verifica se hou-
    # ve a invasão da área do botão para então chamar a função associada à aquele botão.
    def isClicked(self, event):
        x = False
        y = False
        for i in range(len(self.buttons)):
            
            if(event.x >= (self.buttons[i].x + 2) and event.x <= (self.buttons[i].x + self.buttons[i].width - 2)):
                x = True

            if(event.y >= (self.buttons[i].y + 2) and event.y <= (self.buttons[i].y + self.buttons[i].height - 2)):
                y = True

            if((x == True) and (y == True)):
                self.buttons[i].isClicked()
                self.updateDivision()
                self.buttons[i].run()
                time.sleep(0.1)
                self.buttons[i].isOver()
                x = False
                y = False

            else:
                self.buttons[i].isIdle()
                x = False
                y = False

        self.updateDivision()

        return

    # CRIA GRÁFICO
    # 
    # Está função cria um gráfico dentro da divisão.
    def createGraph(self, name, xLabel, yLabel):
        self.graph = Graph(name, xLabel, yLabel, self.frame)
        return

    def removeGraph(self, name):
        self.graph.delPlot(name)
        return

    def clearGraph(self):
        self.graph.clearPlot()
        return

    # MOSTRA O FRAME
    #
    # Essa função coloca o frame no lugar que ele foi criado incialmente.
    def show(self):
        self.frame.place(x=self.xPlace, y=self.yPlace, anchor=NW)
        return

    # ESCONDE O FRAME
    # 
    # Essa funçaõ faz o frame esquecer a posição atual, desaparecendo assim da tela.
    def hide(self):
        self.frame.place_forget()
        return































# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                         [ FUNÇÕES ]                                       |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
#
# Aqui ficam as funções do código, cada função tem uma breve explicação sobre sua funcio-
# nalidade e como ela se comporta.
# 

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                     || MOSTRA AJUDA ||                                    |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# Essa função é responsável por mostrar a divisão de ajuda.
def showHelp():
    global helpDiv
    helpDiv.show()
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      || NOVO FRAME ||                                     |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# Essa função é responsável por criar um novo frame.
def newFrame(s):
    global dados
    global graficos
    global helpDiv

    dados[s].show()
    graficos[s-7].show()

    for i in range(7, 14):
        if(i == s):
            continue
        else:
            dados[i].hide()
            graficos[i-7].hide()

    helpDiv.hide()
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                    || MOSTRA GRÁFICO ||                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# Essa função é responsável por mostrar o gráfico na divisão de gráficos.
def showGraph():
    global graficos
    graficos[-1].show()
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                   || ESCONDE GRÁFICO ||                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# Essa função é responsável por esconder o gráfico na divisão de gráficos.
def hideGraph():
    global graficos
    graficos[-1].hide()
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                              || CALCULA TRAJETÓRIA NUMÉRICA ||                            |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
#
# Essa função realiza o cálcula da trajetória e logo em seguida plota o gráfico da mesma.
def calTrajetoriaNumerica():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[7].entrys[0][1].get())
    b = float(dados[7].entrys[1][1].get())
    c = float(dados[7].entrys[2][1].get())
    angulo = float(dados[7].entrys[3][1].get())
    v0 = float(dados[7].entrys[4][1].get())
    g = float(dados[7].entrys[5][1].get())
    step = float(dados[7].entrys[6][1].get())
    yLimit = float(dados[7].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    x = []
    y = []
    vx = []
    vy = []
    ax = []
    ay = []
    t = []

    vx.append(v0*cos(radians(angulo)))
    vy.append(v0*sin(radians(angulo)))
    ax.append(eq.aXNumerico(c, b, vx[-1], vy[-1]))
    ay.append(eq.aYNumerico(c, b, vx[-1], vy[-1], g))
    x.append(0)
    y.append(0)
    t.append(0)

    stepIncrement = 0
    actualY = 0

    while(actualY >= yLimit and stepIncrement >= 0):

        vx.append(eq.vXNumerico(c, b, angulo, vx[-1], vy[-1], step))
        vy.append(eq.vYNumerico(c, b, angulo, vx[-1], vy[-1], g, step))
        ax.append(eq.aXNumerico(c, b, vx[-1], vy[-1]))
        ay.append(eq.aYNumerico(c, b, vx[-1], vy[-1], g))
        x.append(eq.xNumerico(x[-1], vx[-1], ax[-1], step))
        y.append(eq.yNumerico(y[-1], vy[-1], ay[-1], step))
        stepIncrement += step
        actualY = y[-1]
        t.append(stepIncrement)


    i = 1
    while(name in graficos[0].graph.plotName):
        if name == str(dados[7].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[0].graph.plot(x, y, name)
    graficos[1].graph.plot(t, x, name, "Normal", "No")
    graficos[2].graph.plot(t, y, name, "Normal", "No")
    graficos[3].graph.plot(t, vx, name, "Normal", "No")
    graficos[4].graph.plot(t, vy, name, "Normal", "No")
    graficos[5].graph.plot(t, ax, name, "Normal", "No")
    graficos[6].graph.plot(t, ay, name, "Normal", "No")

    dados[8].entrys[0][1].delete(0, "end")
    dados[8].entrys[0][1].insert(0, name)
    dados[8].entrys[1][1].delete(0, "end")
    dados[8].entrys[1][1].insert(0, b)
    dados[8].entrys[2][1].delete(0, "end")
    dados[8].entrys[2][1].insert(0, c)
    dados[8].entrys[3][1].delete(0, "end")
    dados[8].entrys[3][1].insert(0, angulo)
    dados[8].entrys[4][1].delete(0, "end")
    dados[8].entrys[4][1].insert(0, v0)
    dados[8].entrys[5][1].delete(0, "end")
    dados[8].entrys[5][1].insert(0, g)
    dados[8].entrys[6][1].delete(0, "end")
    dados[8].entrys[6][1].insert(0, step)
    dados[8].entrys[7][1].delete(0, "end")
    dados[8].entrys[7][1].insert(0, ("%.5f" % t[-1]))
    dados[9].entrys[0][1].delete(0, "end")
    dados[9].entrys[0][1].insert(0, name)
    dados[9].entrys[1][1].delete(0, "end")
    dados[9].entrys[1][1].insert(0, b)
    dados[9].entrys[2][1].delete(0, "end")
    dados[9].entrys[2][1].insert(0, c)
    dados[9].entrys[3][1].delete(0, "end")
    dados[9].entrys[3][1].insert(0, angulo)
    dados[9].entrys[4][1].delete(0, "end")
    dados[9].entrys[4][1].insert(0, v0)
    dados[9].entrys[5][1].delete(0, "end")
    dados[9].entrys[5][1].insert(0, g)
    dados[9].entrys[6][1].delete(0, "end")
    dados[9].entrys[6][1].insert(0, step)
    dados[9].entrys[7][1].delete(0, "end")
    dados[9].entrys[7][1].insert(0, ("%.5f" % t[-1]))
    dados[10].entrys[0][1].delete(0, "end")
    dados[10].entrys[0][1].insert(0, name)
    dados[10].entrys[1][1].delete(0, "end")
    dados[10].entrys[1][1].insert(0, b)
    dados[10].entrys[2][1].delete(0, "end")
    dados[10].entrys[2][1].insert(0, c)
    dados[10].entrys[3][1].delete(0, "end")
    dados[10].entrys[3][1].insert(0, angulo)
    dados[10].entrys[4][1].delete(0, "end")
    dados[10].entrys[4][1].insert(0, v0)
    dados[10].entrys[5][1].delete(0, "end")
    dados[10].entrys[5][1].insert(0, g)
    dados[10].entrys[6][1].delete(0, "end")
    dados[10].entrys[6][1].insert(0, step)
    dados[10].entrys[7][1].delete(0, "end")
    dados[10].entrys[7][1].insert(0, ("%.5f" % t[-1]))
    dados[11].entrys[0][1].delete(0, "end")
    dados[11].entrys[0][1].insert(0, name)
    dados[11].entrys[1][1].delete(0, "end")
    dados[11].entrys[1][1].insert(0, b)
    dados[11].entrys[2][1].delete(0, "end")
    dados[11].entrys[2][1].insert(0, c)
    dados[11].entrys[3][1].delete(0, "end")
    dados[11].entrys[3][1].insert(0, angulo)
    dados[11].entrys[4][1].delete(0, "end")
    dados[11].entrys[4][1].insert(0, v0)
    dados[11].entrys[5][1].delete(0, "end")
    dados[11].entrys[5][1].insert(0, g)
    dados[11].entrys[6][1].delete(0, "end")
    dados[11].entrys[6][1].insert(0, step)
    dados[11].entrys[7][1].delete(0, "end")
    dados[11].entrys[7][1].insert(0, ("%.5f" % t[-1]))
    dados[12].entrys[0][1].delete(0, "end")
    dados[12].entrys[0][1].insert(0, name)
    dados[12].entrys[1][1].delete(0, "end")
    dados[12].entrys[1][1].insert(0, b)
    dados[12].entrys[2][1].delete(0, "end")
    dados[12].entrys[2][1].insert(0, c)
    dados[12].entrys[3][1].delete(0, "end")
    dados[12].entrys[3][1].insert(0, angulo)
    dados[12].entrys[4][1].delete(0, "end")
    dados[12].entrys[4][1].insert(0, v0)
    dados[12].entrys[5][1].delete(0, "end")
    dados[12].entrys[5][1].insert(0, g)
    dados[12].entrys[6][1].delete(0, "end")
    dados[12].entrys[6][1].insert(0, step)
    dados[12].entrys[7][1].delete(0, "end")
    dados[12].entrys[7][1].insert(0, ("%.5f" % t[-1]))
    dados[13].entrys[0][1].delete(0, "end")
    dados[13].entrys[0][1].insert(0, name)
    dados[13].entrys[1][1].delete(0, "end")
    dados[13].entrys[1][1].insert(0, b)
    dados[13].entrys[2][1].delete(0, "end")
    dados[13].entrys[2][1].insert(0, c)
    dados[13].entrys[3][1].delete(0, "end")
    dados[13].entrys[3][1].insert(0, angulo)
    dados[13].entrys[4][1].delete(0, "end")
    dados[13].entrys[4][1].insert(0, v0)
    dados[13].entrys[5][1].delete(0, "end")
    dados[13].entrys[5][1].insert(0, g)
    dados[13].entrys[6][1].delete(0, "end")
    dados[13].entrys[6][1].insert(0, step)
    dados[13].entrys[7][1].delete(0, "end")
    dados[13].entrys[7][1].insert(0, ("%.5f" % t[-1]))

    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                             || CALCULA POSIÇÃO x(t) NUMÉRICO ||                           |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calXNumerico():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[8].entrys[0][1].get())
    b = float(dados[8].entrys[1][1].get())
    c = float(dados[8].entrys[2][1].get())
    angulo = float(dados[8].entrys[3][1].get())
    v0 = float(dados[8].entrys[4][1].get())
    g = float(dados[8].entrys[5][1].get())
    step = float(dados[8].entrys[6][1].get())
    maxTime = float(dados[8].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    x = []
    vx = []
    vy = []
    ax = []
    ay = []
    t = []

    vx.append(v0*cos(radians(angulo)))
    vy.append(v0*sin(radians(angulo)))
    ax.append(eq.aXNumerico(c, b, vx[-1], vy[-1]))
    ay.append(eq.aYNumerico(c, b, vx[-1], vy[-1], g))
    x.append(0)
    t.append(0)

    actualTime = 0

    while(actualTime < maxTime):

        vx.append(eq.vXNumerico(c, b, angulo, vx[-1], vy[-1], step))
        vy.append(eq.vYNumerico(c, b, angulo, vx[-1], vy[-1], g, step))
        ax.append(eq.aXNumerico(c, b, vx[-1], vy[-1]))
        ay.append(eq.aYNumerico(c, b, vx[-1], vy[-1], g))
        x.append(eq.xNumerico(x[-1], vx[-1], ax[-1], step))
        actualTime += step
        t.append(actualTime)

    i = 1
    while(name in graficos[1].graph.plotName):
        if name == str(dados[8].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[1].graph.plot(t, x, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                             || CALCULA POSIÇÃO y(t) NUMÉRICO ||                           |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calYNumerico():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[9].entrys[0][1].get())
    b = float(dados[9].entrys[1][1].get())
    c = float(dados[9].entrys[2][1].get())
    angulo = float(dados[9].entrys[3][1].get())
    v0 = float(dados[9].entrys[4][1].get())
    g = float(dados[9].entrys[5][1].get())
    step = float(dados[9].entrys[6][1].get())
    maxTime = float(dados[9].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    y = []
    vx = []
    vy = []
    ax = []
    ay = []
    t = []

    vx.append(v0*cos(radians(angulo)))
    vy.append(v0*sin(radians(angulo)))
    ax.append(eq.aXNumerico(c, b, vx[-1], vy[-1]))
    ay.append(eq.aYNumerico(c, b, vx[-1], vy[-1], g))
    y.append(0)
    t.append(0)

    actualTime = 0

    while(actualTime < maxTime):

        vx.append(eq.vXNumerico(c, b, angulo, vx[-1], vy[-1], step))
        vy.append(eq.vYNumerico(c, b, angulo, vx[-1], vy[-1], g, step))
        ax.append(eq.aXNumerico(c, b, vx[-1], vy[-1]))
        ay.append(eq.aYNumerico(c, b, vx[-1], vy[-1], g))
        y.append(eq.yNumerico(y[-1], vy[-1], ay[-1], step))
        actualTime += step
        t.append(actualTime)

    i = 1
    while(name in graficos[2].graph.plotName):
        if name == str(dados[9].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[2].graph.plot(t, y, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                           || CALCULA VELOCIDADE x(t) NUMÉRICO ||                          |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calVxNumerico():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[10].entrys[0][1].get())
    b = float(dados[10].entrys[1][1].get())
    c = float(dados[10].entrys[2][1].get())
    angulo = float(dados[10].entrys[3][1].get())
    v0 = float(dados[10].entrys[4][1].get())
    g = float(dados[10].entrys[5][1].get())
    step = float(dados[10].entrys[6][1].get())
    maxTime = float(dados[10].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    vx = []
    vy = []
    t = []

    vx.append(v0*cos(radians(angulo)))
    vy.append(v0*sin(radians(angulo)))
    t.append(0)

    actualTime = 0

    while(actualTime < maxTime):

        vx.append(eq.vXNumerico(c, b, angulo, vx[-1], vy[-1], step))
        vy.append(eq.vYNumerico(c, b, angulo, vx[-1], vy[-1], g, step))
        actualTime += step
        t.append(actualTime)

    i = 1
    while(name in graficos[3].graph.plotName):
        if name == str(dados[10].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[3].graph.plot(t, vx, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                           || CALCULA VELOCIDADE y(t) NUMÉRICO ||                          |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calVyNumerico():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[11].entrys[0][1].get())
    b = float(dados[11].entrys[1][1].get())
    c = float(dados[11].entrys[2][1].get())
    angulo = float(dados[11].entrys[3][1].get())
    v0 = float(dados[11].entrys[4][1].get())
    g = float(dados[11].entrys[5][1].get())
    step = float(dados[11].entrys[6][1].get())
    maxTime = float(dados[11].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    vx = []
    vy = []
    t = []

    vx.append(v0*cos(radians(angulo)))
    vy.append(v0*sin(radians(angulo)))
    t.append(0)

    actualTime = 0

    while(actualTime < maxTime):

        vx.append(eq.vXNumerico(c, b, angulo, vx[-1], vy[-1], step))
        vy.append(eq.vYNumerico(c, b, angulo, vx[-1], vy[-1], g, step))
        actualTime += step
        t.append(actualTime)

    i = 1
    while(name in graficos[4].graph.plotName):
        if name == str(dados[11].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[4].graph.plot(t, vy, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                           || CALCULA ACELERAÇÃO x(t) NUMÉRICO ||                          |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calAxNumerico():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[12].entrys[0][1].get())
    b = float(dados[12].entrys[1][1].get())
    c = float(dados[12].entrys[2][1].get())
    angulo = float(dados[12].entrys[3][1].get())
    v0 = float(dados[12].entrys[4][1].get())
    g = float(dados[12].entrys[5][1].get())
    step = float(dados[12].entrys[6][1].get())
    maxTime = float(dados[12].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    vx = []
    vy = []
    ax = []
    t = []

    vx.append(v0*cos(radians(angulo)))
    vy.append(v0*sin(radians(angulo)))
    ax.append(eq.aXNumerico(c, b, vx[-1], vy[-1]))
    t.append(0)

    actualTime = 0

    while(actualTime < maxTime):

        vx.append(eq.vXNumerico(c, b, angulo, vx[-1], vy[-1], step))
        vy.append(eq.vYNumerico(c, b, angulo, vx[-1], vy[-1], g, step))
        ax.append(eq.aXNumerico(c, b, vx[-1], vy[-1]))
        actualTime += step
        t.append(actualTime)

    i = 1
    while(name in graficos[5].graph.plotName):
        if name == str(dados[12].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[5].graph.plot(t, ax, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                           || CALCULA ACELERAÇÃO y(t) NUMÉRICO ||                          |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calAyNumerico():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[13].entrys[0][1].get())
    b = float(dados[13].entrys[1][1].get())
    c = float(dados[13].entrys[2][1].get())
    angulo = float(dados[13].entrys[3][1].get())
    v0 = float(dados[13].entrys[4][1].get())
    g = float(dados[13].entrys[5][1].get())
    step = float(dados[13].entrys[6][1].get())
    maxTime = float(dados[13].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    vx = []
    vy = []
    ay = []
    t = []

    vx.append(v0*cos(radians(angulo)))
    vy.append(v0*sin(radians(angulo)))
    ay.append(eq.aYNumerico(c, b, vx[-1], vy[-1], g))
    t.append(0)

    actualTime = 0

    while(actualTime < maxTime):

        vx.append(eq.vXNumerico(c, b, angulo, vx[-1], vy[-1], step))
        vy.append(eq.vYNumerico(c, b, angulo, vx[-1], vy[-1], g, step))
        ay.append(eq.aYNumerico(c, b, vx[-1], vy[-1], g))
        actualTime += step
        t.append(actualTime)

    i = 1
    while(name in graficos[6].graph.plotName):
        if name == str(dados[13].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[6].graph.plot(t, y, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                     || CALCULA BETA ||                                    |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calculaBeta():
    global calculadoraBeta

    eq = Equacoes()

    n = float(calculadoraBeta.entrys[0][1].get())
    r = float(calculadoraBeta.entrys[1][1].get())
    m = float(calculadoraBeta.entrys[2][1].get())

    beta = eq.bCalculator(n, r, m)

    calculadoraBeta.entrys[3][1].config(text=str("%.5f" % beta))
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                     || CALCULA GAMMA ||                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calculaGamma():
    global calculadoraGamma

    eq = Equacoes()

    C = float(calculadoraGamma.entrys[0][1].get())
    p = float(calculadoraGamma.entrys[1][1].get())
    r = float(calculadoraGamma.entrys[2][1].get())
    m = float(calculadoraGamma.entrys[3][1].get())

    gamma = eq.cCalculator(C, p, r, m)

    calculadoraGamma.entrys[4][1].config(text=str("%.5f" % gamma))
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                  || CALCULA TRAJETÓRIA ||                                 |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
#
# Essa função realiza o cálcula da trajetória e logo em seguida plota o gráfico da mesma.
def calTrajetoria():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[0].entrys[0][1].get())
    beta = float(dados[0].entrys[1][1].get())
    theta = float(dados[0].entrys[2][1].get())
    v0 = float(dados[0].entrys[3][1].get())
    g = float(dados[0].entrys[4][1].get())
    step = float(dados[0].entrys[5][1].get())
    yLimit = float(dados[0].entrys[6][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    x = []
    y = []

    stepIncrement = 0
    last = 0

    while(last >= yLimit):
        if (beta != 0 and theta != 90 and theta != 270):
            y.append(eq.trajetoria(beta, theta, v0, stepIncrement, g))
            x.append(stepIncrement)
        elif (beta != 0 and (theta == 90 or theta == 270)):
            y.append(eq.y(beta, theta, v0, stepIncrement, g))
            x.append(0)
        elif (beta == 0 and (theta == 90 or theta == 270)):
            y.append(eq.yDef(theta, v0, stepIncrement, g))
            x.append(0)
        else:
            y.append(eq.trajetoriaDef(theta, v0, stepIncrement, g))
            x.append(stepIncrement)
        last = y[-1]
        if(theta > 90 and theta < 270 ):
            stepIncrement -= step
        else:
            stepIncrement += step

    i = 1
    while(name in graficos[0].graph.plotName):
        if name == str(dados[0].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[0].graph.plot(x, y, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                  || CALCULA POSIÇÃO x(t) ||                               |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calX():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[1].entrys[0][1].get())
    beta = float(dados[1].entrys[1][1].get())
    theta = float(dados[1].entrys[2][1].get())
    v0 = float(dados[1].entrys[3][1].get())
    t0 = float(dados[1].entrys[4][1].get())
    t1 = float(dados[1].entrys[5][1].get())
    step = float(dados[1].entrys[6][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    if(t0 < 0):
        print("Por favor, insira um t0 válido!")
        return

    if(t1 < t0):
        print("Por favor, insira um t1 válido!")
        return

    x = []
    y = []

    stepIncrement = 0

    while(t0 <= t1):
        if(beta != 0):
            y.append(eq.x(beta, theta, v0, t0))
        else:
            y.append(eq.xDef(theta, v0, t0))
        x.append(t0)
        t0 += step

    i = 1
    while(name in graficos[1].graph.plotName):
        if name == str(dados[1].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[1].graph.plot(x, y, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                  || CALCULA POSIÇÃO y(t) ||                               |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calY():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[2].entrys[0][1].get())
    beta = float(dados[2].entrys[1][1].get())
    theta = float(dados[2].entrys[2][1].get())
    v0 = float(dados[2].entrys[3][1].get())
    g = float(dados[2].entrys[4][1].get())
    t0 = float(dados[2].entrys[5][1].get())
    t1 = float(dados[2].entrys[6][1].get())
    step = float(dados[2].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    if(t0 < 0):
        print("Por favor, insira um t0 válido!")
        return

    if(t1 < t0):
        print("Por favor, insira um t1 válido!")
        return

    x = []
    y = []

    stepIncrement = 0

    while(t0 <= t1):
        if(beta != 0):
            y.append(eq.y(beta, theta, v0, t0, g))
        else:
            y.append(eq.yDef(theta, v0, t0, g))
        x.append(t0)
        t0 += step

    i = 1
    while(name in graficos[2].graph.plotName):
        if name == str(dados[2].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[2].graph.plot(x, y, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                || CALCULA VELOCIDADE x(t) ||                              |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calVx():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[3].entrys[0][1].get())
    beta = float(dados[3].entrys[1][1].get())
    theta = float(dados[3].entrys[2][1].get())
    v0 = float(dados[3].entrys[3][1].get())
    t0 = float(dados[3].entrys[4][1].get())
    t1 = float(dados[3].entrys[5][1].get())
    step = float(dados[3].entrys[6][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    if(t0 < 0):
        print("Por favor, insira um t0 válido!")
        return

    if(t1 < t0):
        print("Por favor, insira um t1 válido!")
        return

    x = []
    y = []

    stepIncrement = 0

    while(t0 <= t1):
        if(beta != 0):
            y.append(eq.vX(beta, theta, v0, t0))
        else:
            y.append(eq.vXDef(theta, v0, t0))
        x.append(t0)
        t0 += step

    i = 1
    while(name in graficos[3].graph.plotName):
        if name == str(dados[3].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[3].graph.plot(x, y, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                || CALCULA VELOCIDADE y(t) ||                              |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calVy():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[4].entrys[0][1].get())
    beta = float(dados[4].entrys[1][1].get())
    theta = float(dados[4].entrys[2][1].get())
    v0 = float(dados[4].entrys[3][1].get())
    g = float(dados[4].entrys[4][1].get())
    t0 = float(dados[4].entrys[5][1].get())
    t1 = float(dados[4].entrys[6][1].get())
    step = float(dados[4].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    if(t0 < 0):
        print("Por favor, insira um t0 válido!")
        return

    if(t1 < t0):
        print("Por favor, insira um t1 válido!")
        return

    x = []
    y = []

    stepIncrement = 0

    while(t0 <= t1):
        if(beta != 0):
            y.append(eq.vY(beta, theta, v0, t0, g))
        else:
            y.append(eq.vYDef(theta, v0, t0, g))
        x.append(t0)
        t0 += step

    i = 1
    while(name in graficos[4].graph.plotName):
        if name == str(dados[4].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[4].graph.plot(x, y, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                || CALCULA ACELERAÇÃO x(t) ||                              |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calAx():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[5].entrys[0][1].get())
    beta = float(dados[5].entrys[1][1].get())
    theta = float(dados[5].entrys[2][1].get())
    v0 = float(dados[5].entrys[3][1].get())
    t0 = float(dados[5].entrys[4][1].get())
    t1 = float(dados[5].entrys[5][1].get())
    step = float(dados[5].entrys[6][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    if(t0 < 0):
        print("Por favor, insira um t0 válido!")
        return

    if(t1 < t0):
        print("Por favor, insira um t1 válido!")
        return

    x = []
    y = []

    stepIncrement = 0

    while(t0 <= t1):
        if(beta != 0):
            y.append(eq.aX(beta, theta, v0, t0))
        else:
            y.append(eq.aXDef())
        x.append(t0)
        t0 += step

    i = 1
    while(name in graficos[5].graph.plotName):
        if name == str(dados[5].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[5].graph.plot(x, y, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                || CALCULA ACELERAÇÃO y(t) ||                              |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def calAy():
    global graficos
    global dados

    eq = Equacoes()

    name = str(dados[6].entrys[0][1].get())
    beta = float(dados[6].entrys[1][1].get())
    theta = float(dados[6].entrys[2][1].get())
    v0 = float(dados[6].entrys[3][1].get())
    g = float(dados[6].entrys[4][1].get())
    t0 = float(dados[6].entrys[5][1].get())
    t1 = float(dados[6].entrys[6][1].get())
    step = float(dados[6].entrys[7][1].get())

    if(name == ""):
        print("Por favor, coloque um nome para seu plot!")
        return

    if(t0 < 0):
        print("Por favor, insira um t0 válido!")
        return

    if(t1 < t0):
        print("Por favor, insira um t1 válido!")
        return

    x = []
    y = []

    stepIncrement = 0

    while(t0 <= t1):
        if(beta != 0):
            y.append(eq.aY(beta, theta, v0, t0, g))
        else:
            y.append(eq.aYDef(g))
        x.append(t0)
        t0 += step

    i = 1
    while(name in graficos[6].graph.plotName):
        if name == str(dados[6].entrys[0][1].get()):
            name = name + "(" + str(i) + ")"
        else:
            name = name.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1

    graficos[6].graph.plot(x, y, name)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                    || REMOVE UM PLOT ||                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def delPlot(divIndex):
    global graficos
    if(divIndex > 6):
        graphIndex = divIndex - 7
    else:
        graphIndex = divIndex
    plotName = str(dados[divIndex].entrys[0][1].get())
    graficos[graphIndex].removeGraph(plotName)
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                   || LIMPA O GRÁFICO ||                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def clearPlot(divIndex):
    global graficos
    graficos[divIndex].clearGraph()
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                            || IMPORTA VALORES EXPERIMENTAIS ||                            |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def importPlot(actual):
    global graficos
    filePath = csvie.getFile()
    line = csvie.getFileLines(filePath)
    file = filePath.split("/")[-1]
    fileBackup = file
    i = 1

    while(file in graficos[0].graph.plotName):
        if file == fileBackup:
            file = file + "(" + str(i) + ")"
        else:
            file = file.replace(("(" + str((i-1)) + ")"), ("(" + str(i) + ")"))
        i += 1
    
    if(line[1] != [] and line[2] != []):
        if (actual == 0):
            graficos[0].graph.plot(line[1], line[2], file, "Experimental")
        else:
            graficos[0].graph.plot(line[1], line[2], file, "Experimental", "No")
    
    if(line[0] != [] and line[1] != []):
        if (actual == 1):
            graficos[1].graph.plot(line[0], line[1], file, "Experimental")
        else:
            graficos[1].graph.plot(line[0], line[1], file, "Experimental", "No")
    
    if(line[0] != [] and line[2] != []):
        if (actual == 2):
            graficos[2].graph.plot(line[0], line[2], file, "Experimental")
        else:
            graficos[2].graph.plot(line[0], line[2], file, "Experimental", "No")
    
    if(line[0][1:-1] != [] and line[3] != []):
        if (actual == 3):
            graficos[3].graph.plot(line[0][1:-1], line[3], file, "Experimental")
        else:
            graficos[3].graph.plot(line[0][1:-1], line[3], file, "Experimental", "No")
    
    if(line[0][1:-1] != [] and line[4] != []):
        if (actual == 4):
            graficos[4].graph.plot(line[0][1:-1], line[4], file, "Experimental")
        else:
            graficos[4].graph.plot(line[0][1:-1], line[4], file, "Experimental", "No")
    
    if(line[0][2:-2] != [] and line[5] != []):
        if (actual == 5):
            graficos[5].graph.plot(line[0][2:-2], line[5], file, "Experimental")
        else:
            graficos[5].graph.plot(line[0][2:-2], line[5], file, "Experimental", "No")
    
    if(line[0][2:-2] != [] and line[6] != []):
        if (actual == 6):
            graficos[6].graph.plot(line[0][2:-2], line[6], file, "Experimental")
        else:
            graficos[6].graph.plot(line[0][2:-2], line[6], file, "Experimental", "No")
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                || EXPORTA UM ÚNICO PLOT ||                                |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def exportSinglePlot(i):
    global graficos
    filename = str(dados[i].entrys[0][1].get())
    
    if(i > 6):
        graphIndex = i - 7
    else:
        graphIndex = i
    try:
        plot = graficos[graphIndex].graph.plotName.index(filename)
    except:
        print("Plot not found!")
        return
    v = [graficos[graphIndex].graph.xPoint[plot], graficos[graphIndex].graph.yPoint[plot]]
    x = graficos[graphIndex].graph.xLabel
    y = graficos[graphIndex].graph.yLabel
    csvie.exportSingleGraph((filename + "_" + graficos[graphIndex].graph.graphName.split()[0]), v, x, y)
    print("Plot exported with success!")
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                   || MÉTODO NUMÉRICO ||                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def changeNumeric(i):
    global dados
    dados[i].hide()
    dados[i+7].show()
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                   || MÉTODO ANALÍTCO ||                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def changeAnalitic(i):
    global dados
    dados[i].hide()
    dados[i-7].show()
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                               || LIMPA TODOS OS GRÁFICOS ||                               |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def clearAllGraphs():
    global graficos
    graficos[0].clearGraph()
    graficos[1].clearGraph()
    graficos[2].clearGraph()
    graficos[3].clearGraph()
    graficos[4].clearGraph()
    graficos[5].clearGraph()
    graficos[6].clearGraph()
    return

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                   || ACESSA O GITHUB ||                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

def gitHubOpen():
    url = "https://github.com/Zakonildo/FEI-Projectile-Motion-Simulation"
    webbrowser.open_new(url)
    return































# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                         [ MAIN ]                                          |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# Aqui é o local em que serão feitas as chamadas de funções e declarações de classes e de
# variáveis.

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                        || ROOT ||                                         |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
# 
# O root é a nossa janela completa, ele é o responsável pelo tamanho, título, e ícone da
# janela do Software. É dentro do root que colocaremos as nossas divisões.

root = Tk()
root.title("FEI - Projectile Motion Simulation")
root.iconphoto(True, PhotoImage(file=os.getcwd()+"\\Icon.png"))

positionRight = int(root.winfo_screenwidth()/2 - 1280/2)
positionDown = int(root.winfo_screenheight()/2 - 720/2)

root.resizable(width=False, height=False)
root.geometry("1280x720" "+{}+{}".format(positionRight, positionDown))

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                  INICIALIZA AS DIVISÕES                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

menu = Division(root, 1280/4, 720/1.5, 0, 0, "Background_Pattern1.png")
calculadoraBeta = Division(root , 1280/4, 720/3, 0, 720/1.5, "Background_Beta.png")
calculadoraGamma = Division(root , 1280/4, 720/3, 1280/4, 720/1.5, "Background_Gamma.png")
simulacao = Division(root, 1280/2, 720/2, 1280/2, 720/2, "Animation_Background.png")

dados = []
graficos = []

for i in range(0, 14):
    dados.append(Division(root, 1280/4, 720/1.5, 1280/4, 0, "Background_Pattern" + str(1) + ".png"))

    if(i != 7):
        dados[-1].hide()

for i in range(0, 7):
    graficos.append(Division(root, 1280/2, 720/2, 1280/2, 0, "#2975A3"))
    if(i != 0):
        graficos[-1].hide()

helpDiv = Division(root, 1280/4, 720/1.5, 1280/4, 0, "Background_PatternHelp.png")
helpDiv.hide()

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                   FUNÇÕES PARA BOTÕES                                     |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

# [0]      lambda s=0 : newFrame(s),
# [1]      lambda s=1 : newFrame(s),
# [2]      lambda s=2 : newFrame(s),
# [3]      lambda s=3 : newFrame(s),
# [4]      lambda s=4 : newFrame(s),
# [5]      lambda s=5 : newFrame(s),
# [6]      lambda s=6 : newFrame(s),
# [7]      lambda : hideGraph(),
# [8]      lambda : calTrajetoria(),
# [9]      lambda : calX(),
# [10]     lambda : calY(),
# [11]     lambda : calVx(),
# [12]     lambda : calVy(),
# [13]     lambda : calAx(),
# [14]     lambda : calAy(),
# [15]     lambda divIndex=0 : delPlot(divIndex),
# [16]     lambda divIndex=1 : delPlot(divIndex),
# [17]     lambda divIndex=2 : delPlot(divIndex),
# [18]     lambda divIndex=3 : delPlot(divIndex),
# [19]     lambda divIndex=4 : delPlot(divIndex),
# [20]     lambda divIndex=5 : delPlot(divIndex),
# [21]     lambda divIndex=6 : delPlot(divIndex),
# [22]     lambda divIndex=0 : clearPlot(divIndex),
# [23]     lambda divIndex=1 : clearPlot(divIndex),
# [24]     lambda divIndex=2 : clearPlot(divIndex),
# [25]     lambda divIndex=3 : clearPlot(divIndex),
# [26]     lambda divIndex=4 : clearPlot(divIndex),
# [27]     lambda divIndex=5 : clearPlot(divIndex),
# [28]     lambda divIndex=6 : clearPlot(divIndex),
# [29]     lambda actual=-1 : importPlot(actual),
# [30]     lambda divIndex=0 : exportSinglePlot(divIndex),
# [31]     lambda divIndex=1 : exportSinglePlot(divIndex),
# [32]     lambda divIndex=2 : exportSinglePlot(divIndex),
# [33]     lambda divIndex=3 : exportSinglePlot(divIndex),
# [34]     lambda divIndex=4 : exportSinglePlot(divIndex),
# [35]     lambda divIndex=5 : exportSinglePlot(divIndex),
# [36]     lambda divIndex=6 : exportSinglePlot(divIndex),
# [37]     lambda divIndex=0 : changeNumeric(divIndex),
# [38]     lambda divIndex=1 : changeNumeric(divIndex),
# [39]     lambda divIndex=2 : changeNumeric(divIndex),
# [40]     lambda divIndex=3 : changeNumeric(divIndex),
# [41]     lambda divIndex=4 : changeNumeric(divIndex),
# [42]     lambda divIndex=5 : changeNumeric(divIndex),
# [43]     lambda divIndex=6 : changeNumeric(divIndex),
# [44]     lambda divIndex=7 : changeAnalitic(divIndex),
# [45]     lambda divIndex=8 : changeAnalitic(divIndex),
# [46]     lambda divIndex=9 : changeAnalitic(divIndex),
# [47]     lambda divIndex=10 : changeAnalitic(divIndex),
# [48]     lambda divIndex=11 : changeAnalitic(divIndex),
# [49]     lambda divIndex=12 : changeAnalitic(divIndex),
# [50]     lambda divIndex=13 : changeAnalitic(divIndex),
# [51]     lambda : calTrajetoriaNumerica(),
# [52]     lambda : calXNumerico(),
# [53]     lambda : calYNumerico(),
# [54]     lambda : calVxNumerico(),
# [55]     lambda : calVyNumerico(),
# [56]     lambda : calAxNumerico(),
# [57]     lambda : calAyNumerico()
# [58]     lambda : calculaBeta(),
# [59]     lambda : calculaGamma(),
# [60]     lambda divIndex=7 : delPlot(divIndex),
# [61]     lambda divIndex=8 : delPlot(divIndex),
# [62]     lambda divIndex=9 : delPlot(divIndex),
# [63]     lambda divIndex=10 : delPlot(divIndex),
# [64]     lambda divIndex=11 : delPlot(divIndex),
# [65]     lambda divIndex=12 : delPlot(divIndex),
# [66]     lambda divIndex=13 : delPlot(divIndex),
# [67]     lambda actual=0 : importPlot(actual),
# [68]     lambda actual=1 : importPlot(actual),
# [69]     lambda actual=2 : importPlot(actual),
# [70]     lambda actual=3 : importPlot(actual),
# [71]     lambda actual=4 : importPlot(actual),
# [72]     lambda actual=5 : importPlot(actual),
# [73]     lambda actual=6 : importPlot(actual),
# [74]     lambda divIndex=7 : exportSinglePlot(divIndex),
# [75]     lambda divIndex=8 : exportSinglePlot(divIndex),
# [76]     lambda divIndex=9 : exportSinglePlot(divIndex),
# [77]     lambda divIndex=10 : exportSinglePlot(divIndex),
# [78]     lambda divIndex=11 : exportSinglePlot(divIndex),
# [79]     lambda divIndex=12 : exportSinglePlot(divIndex),
# [80]     lambda divIndex=13 : exportSinglePlot(divIndex),
# [81]     lambda : clearAllGraphs(),
# [82]     lambda : gitHubOpen(),

f = [lambda s= (0+7) : newFrame(s),
     lambda s= (1+7) : newFrame(s),
     lambda s= (2+7) : newFrame(s),
     lambda s= (3+7) : newFrame(s),
     lambda s= (4+7) : newFrame(s),
     lambda s= (5+7) : newFrame(s),
     lambda s= (6+7) : newFrame(s),
     lambda : showHelp(),
     lambda : calTrajetoria(),
     lambda : calX(),
     lambda : calY(),
     lambda : calVx(),
     lambda : calVy(),
     lambda : calAx(),
     lambda : calAy(),
     lambda divIndex=0 : delPlot(divIndex),
     lambda divIndex=1 : delPlot(divIndex),
     lambda divIndex=2 : delPlot(divIndex),
     lambda divIndex=3 : delPlot(divIndex),
     lambda divIndex=4 : delPlot(divIndex),
     lambda divIndex=5 : delPlot(divIndex),
     lambda divIndex=6 : delPlot(divIndex),
     lambda divIndex=0 : clearPlot(divIndex),
     lambda divIndex=1 : clearPlot(divIndex),
     lambda divIndex=2 : clearPlot(divIndex),
     lambda divIndex=3 : clearPlot(divIndex),
     lambda divIndex=4 : clearPlot(divIndex),
     lambda divIndex=5 : clearPlot(divIndex),
     lambda divIndex=6 : clearPlot(divIndex),
     lambda actual=-1 : importPlot(actual),
     lambda divIndex=0 : exportSinglePlot(divIndex),
     lambda divIndex=1 : exportSinglePlot(divIndex),
     lambda divIndex=2 : exportSinglePlot(divIndex),
     lambda divIndex=3 : exportSinglePlot(divIndex),
     lambda divIndex=4 : exportSinglePlot(divIndex),
     lambda divIndex=5 : exportSinglePlot(divIndex),
     lambda divIndex=6 : exportSinglePlot(divIndex),
     lambda divIndex=0 : changeNumeric(divIndex),
     lambda divIndex=1 : changeNumeric(divIndex),
     lambda divIndex=2 : changeNumeric(divIndex),
     lambda divIndex=3 : changeNumeric(divIndex),
     lambda divIndex=4 : changeNumeric(divIndex),
     lambda divIndex=5 : changeNumeric(divIndex),
     lambda divIndex=6 : changeNumeric(divIndex),
     lambda divIndex=7 : changeAnalitic(divIndex),
     lambda divIndex=8 : changeAnalitic(divIndex),
     lambda divIndex=9 : changeAnalitic(divIndex),
     lambda divIndex=10 : changeAnalitic(divIndex),
     lambda divIndex=11 : changeAnalitic(divIndex),
     lambda divIndex=12 : changeAnalitic(divIndex),
     lambda divIndex=13 : changeAnalitic(divIndex),
     lambda : calTrajetoriaNumerica(),
     lambda : calXNumerico(),
     lambda : calYNumerico(),
     lambda : calVxNumerico(),
     lambda : calVyNumerico(),
     lambda : calAxNumerico(),
     lambda : calAyNumerico(),
     lambda : calculaBeta(),
     lambda : calculaGamma(),
     lambda divIndex=7 : delPlot(divIndex),
     lambda divIndex=8 : delPlot(divIndex),
     lambda divIndex=9 : delPlot(divIndex),
     lambda divIndex=10 : delPlot(divIndex),
     lambda divIndex=11 : delPlot(divIndex),
     lambda divIndex=12 : delPlot(divIndex),
     lambda divIndex=13 : delPlot(divIndex),
     lambda actual=0 : importPlot(actual),
     lambda actual=1 : importPlot(actual),
     lambda actual=2 : importPlot(actual),
     lambda actual=3 : importPlot(actual),
     lambda actual=4 : importPlot(actual),
     lambda actual=5 : importPlot(actual),
     lambda actual=6 : importPlot(actual),
     lambda divIndex=7 : exportSinglePlot(divIndex),
     lambda divIndex=8 : exportSinglePlot(divIndex),
     lambda divIndex=9 : exportSinglePlot(divIndex),
     lambda divIndex=10 : exportSinglePlot(divIndex),
     lambda divIndex=11 : exportSinglePlot(divIndex),
     lambda divIndex=12 : exportSinglePlot(divIndex),
     lambda divIndex=13 : exportSinglePlot(divIndex),
     lambda : clearAllGraphs(),
     lambda : gitHubOpen(),]

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      CONFIGURAÇÕES                                        |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

helpDiv.newButton(
    (helpDiv.divWidth/2 - 70), 
    25, 
    140, 
    50, 
    "ClearAll", 
    f[81])

helpDiv.newButton(
    (helpDiv.divWidth/2 - 39), 
    315, 
    78, 
    78, 
    "GitHub", 
    f[82])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      BOTÕES DO MENU                                       |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

menu.newButton(0, 0, 320, 60, "Button01", f[0])
menu.newButton(0, 60, 320, 60, "Button02", f[1])
menu.newButton(0, 120, 320, 60, "Button03", f[2])
menu.newButton(0, 180, 320, 60, "Button04", f[3])
menu.newButton(0, 240, 320, 60, "Button05", f[4])
menu.newButton(0, 300, 320, 60, "Button06", f[5])
menu.newButton(0, 360, 320, 60, "Button07", f[6])
menu.newButton(0, 420, 320, 60, "Button08", f[7])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                        TRAJETÓRIA                                         |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[0].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[0].newLabel("Beta", 17, 15, 35, "Beta.png", True)
dados[0].newLabel("Theta", 17, 15, 70, "Theta.png", True)
dados[0].newLabel("v0", 17, 15, 105, "v0.png", True)
dados[0].newLabel("Gravidade", 17, 15, 140, "Gravidade.png", True)
dados[0].newLabel("Passo", 14, 13, 175, "Passo.png", True)
dados[0].newLabel("Limite", 7, 13, 210, "Limite.png", True)

dados[0].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[0].newLabel("Unidade02", 17, 265, 70, "GrausU02.png", False)
dados[0].newLabel("Unidade03", 17, 265, 105, "VelU03.png", False)
dados[0].newLabel("Unidade04", 17, 265, 140, "GravU04.png", False)
dados[0].newLabel("Unidade05", 17, 265, 175, "MetroU05.png", False)
dados[0].newLabel("Unidade06", 17, 265, 210, "MetroU06.png", False)
dados[0].newLabel("MetodoAna", 17, 110, 277, "None", False)

dados[0].newButton(((dados[0].divWidth/2 + 75) - 60),
                    ((dados[0].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[22])

dados[0].newButton(((dados[0].divWidth/2 + 75) - 60),
                    ((dados[0].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[15])

dados[0].newButton(((dados[0].divWidth/2 - 75) - 60),
                    ((dados[0].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[67])

dados[0].newButton(((dados[0].divWidth/2 - 75) - 60),
                    ((dados[0].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[8])

dados[0].newButton(((dados[0].divWidth/2 - 75) - 60),
                    ((dados[0].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[30])

dados[0].newButton(((dados[0].divWidth/2 + 75) - 60),
                    ((dados[0].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodNum",
                    f[37])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                       POSIÇÃO x(t)                                        |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[1].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[1].newLabel("Beta", 17, 15, 35, "Beta.png", True)
dados[1].newLabel("Theta", 17, 15, 70, "Theta.png", True)
dados[1].newLabel("v0", 17, 15, 105, "v0.png", True)
dados[1].newLabel("t0", 17, 15, 140, "t0_x.png", True)
dados[1].newLabel("t1", 17, 15, 175, "t1_x.png", True)
dados[1].newLabel("Passo", 14, 13, 210, "Passo_x.png", True)

dados[1].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[1].newLabel("Unidade02", 17, 265, 70, "GrausU02.png", False)
dados[1].newLabel("Unidade03", 17, 265, 105, "VelU03.png", False)
dados[1].newLabel("Unidade04", 17, 265, 140, "TimeU04.png", False)
dados[1].newLabel("Unidade05", 17, 265, 175, "TimeU05.png", False)
dados[1].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[1].newLabel("MetodoAna", 17, 110, 277, "None", False)


dados[1].newButton(((dados[1].divWidth/2 + 75) - 60),
                    ((dados[1].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodNum",
                    f[38])

dados[1].newButton(((dados[1].divWidth/2 + 75) - 60),
                    ((dados[1].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[23])

dados[1].newButton(((dados[1].divWidth/2 + 75) - 60),
                    ((dados[1].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[16])

dados[1].newButton(((dados[1].divWidth/2 - 75) - 60),
                    ((dados[1].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[31])

dados[1].newButton(((dados[1].divWidth/2 - 75) - 60),
                    ((dados[1].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[68])

dados[1].newButton(((dados[1].divWidth/2 - 75) - 60),
                    ((dados[1].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[9])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                       POSIÇÃO y(t)                                        |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[2].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[2].newLabel("Beta", 17, 15, 35, "Beta.png", True)
dados[2].newLabel("Theta", 17, 15, 70, "Theta.png", True)
dados[2].newLabel("v0", 17, 15, 105, "v0.png", True)
dados[2].newLabel("Gravidade", 17, 15, 140, "Gravidade.png", True)
dados[2].newLabel("t0", 17, 15, 175, "t0_y.png", True)
dados[2].newLabel("t1", 17, 15, 210, "t1_y.png", True)
dados[2].newLabel("Passo", 14, 13, 245, "Passo_y.png", True)
dados[2].newLabel("MetodoAna", 17, 110, 277, "None", False)

dados[2].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[2].newLabel("Unidade02", 17, 265, 70, "GrausU02.png", False)
dados[2].newLabel("Unidade03", 17, 265, 105, "VelU03.png", False)
dados[2].newLabel("Unidade04", 17, 265, 140, "GravU04.png", False)
dados[2].newLabel("Unidade05", 17, 265, 175, "TimeU05.png", False)
dados[2].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[2].newLabel("Unidade07", 17, 265, 245, "TimeU07.png", False)

dados[2].newButton(((dados[2].divWidth/2 + 75) - 60),
                    ((dados[2].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodNum",
                    f[39])

dados[2].newButton(((dados[2].divWidth/2 + 75) - 60),
                    ((dados[2].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[24])

dados[2].newButton(((dados[2].divWidth/2 - 75) - 60),
                    ((dados[2].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[69])

dados[2].newButton(((dados[2].divWidth/2 + 75) - 60),
                    ((dados[2].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[17])

dados[2].newButton(((dados[2].divWidth/2 - 75) - 60),
                    ((dados[2].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[10])

dados[2].newButton(((dados[2].divWidth/2 - 75) - 60),
                    ((dados[2].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[32])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      VELOCIDADE x(t)                                      |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[3].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[3].newLabel("Beta", 17, 15, 35, "Beta.png", True)
dados[3].newLabel("Theta", 17, 15, 70, "Theta.png", True)
dados[3].newLabel("v0", 17, 15, 105, "v0.png", True)
dados[3].newLabel("t0", 17, 15, 140, "t0_x.png", True)
dados[3].newLabel("t1", 17, 15, 175, "t1_x.png", True)
dados[3].newLabel("Passo", 14, 13, 210, "Passo_x.png", True)

dados[3].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[3].newLabel("Unidade02", 17, 265, 70, "GrausU02.png", False)
dados[3].newLabel("Unidade03", 17, 265, 105, "VelU03.png", False)
dados[3].newLabel("Unidade04", 17, 265, 140, "TimeU04.png", False)
dados[3].newLabel("Unidade05", 17, 265, 175, "TimeU05.png", False)
dados[3].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[3].newLabel("MetodoAna", 17, 110, 277, "None", False)

dados[3].newButton(((dados[3].divWidth/2 + 75) - 60),
                    ((dados[3].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodNum",
                    f[40])

dados[3].newButton(((dados[3].divWidth/2 + 75) - 60),
                    ((dados[3].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[25])

dados[3].newButton(((dados[3].divWidth/2 + 75) - 60),
                    ((dados[3].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[18])

dados[3].newButton(((dados[3].divWidth/2 - 75) - 60),
                    ((dados[3].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[33])

dados[3].newButton(((dados[3].divWidth/2 - 75) - 60),
                    ((dados[3].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[70])

dados[3].newButton(((dados[3].divWidth/2 - 75) - 60),
                    ((dados[3].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[11])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      VELOCIDADE y(t)                                      |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[4].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[4].newLabel("Beta", 17, 15, 35, "Beta.png", True)
dados[4].newLabel("Theta", 17, 15, 70, "Theta.png", True)
dados[4].newLabel("v0", 17, 15, 105, "v0.png", True)
dados[4].newLabel("Gravidade", 17, 15, 140, "Gravidade.png", True)
dados[4].newLabel("t0", 17, 15, 175, "t0_y.png", True)
dados[4].newLabel("t1", 17, 15, 210, "t1_y.png", True)
dados[4].newLabel("Passo", 14, 13, 245, "Passo_y.png", True)

dados[4].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[4].newLabel("Unidade02", 17, 265, 70, "GrausU02.png", False)
dados[4].newLabel("Unidade03", 17, 265, 105, "VelU03.png", False)
dados[4].newLabel("Unidade04", 17, 265, 140, "GravU04.png", False)
dados[4].newLabel("Unidade05", 17, 265, 175, "TimeU05.png", False)
dados[4].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[4].newLabel("Unidade07", 17, 265, 245, "TimeU07.png", False)
dados[4].newLabel("MetodoAna", 17, 110, 277, "None", False)

dados[4].newButton(((dados[4].divWidth/2 + 75) - 60),
                    ((dados[4].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodNum",
                    f[41])

dados[4].newButton(((dados[4].divWidth/2 + 75) - 60),
                    ((dados[4].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[26])

dados[4].newButton(((dados[4].divWidth/2 - 75) - 60),
                    ((dados[4].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[71])

dados[4].newButton(((dados[4].divWidth/2 + 75) - 60),
                    ((dados[4].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[19])

dados[4].newButton(((dados[4].divWidth/2 - 75) - 60),
                    ((dados[4].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[12])

dados[4].newButton(((dados[4].divWidth/2 - 75) - 60),
                    ((dados[4].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[34])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      ACELERAÇÃO x(t)                                      |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[5].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[5].newLabel("Beta", 17, 15, 35, "Beta.png", True)
dados[5].newLabel("Theta", 17, 15, 70, "Theta.png", True)
dados[5].newLabel("v0", 17, 15, 105, "v0.png", True)
dados[5].newLabel("t0", 17, 15, 140, "t0_x.png", True)
dados[5].newLabel("t1", 17, 15, 175, "t1_x.png", True)
dados[5].newLabel("Passo", 14, 13, 210, "Passo_x.png", True)

dados[5].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[5].newLabel("Unidade02", 17, 265, 70, "GrausU02.png", False)
dados[5].newLabel("Unidade03", 17, 265, 105, "VelU03.png", False)
dados[5].newLabel("Unidade04", 17, 265, 140, "TimeU04.png", False)
dados[5].newLabel("Unidade05", 17, 265, 175, "TimeU05.png", False)
dados[5].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[5].newLabel("MetodoAna", 17, 110, 277, "None", False)

dados[5].newButton(((dados[5].divWidth/2 + 75) - 60),
                    ((dados[5].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodNum",
                    f[42])

dados[5].newButton(((dados[5].divWidth/2 + 75) - 60),
                    ((dados[5].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[27])

dados[5].newButton(((dados[5].divWidth/2 - 75) - 60),
                    ((dados[5].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[72])

dados[5].newButton(((dados[5].divWidth/2 + 75) - 60),
                    ((dados[5].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[20])

dados[5].newButton(((dados[5].divWidth/2 - 75) - 60),
                    ((dados[5].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[13])

dados[5].newButton(((dados[5].divWidth/2 - 75) - 60),
                    ((dados[5].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[35])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      ACELERAÇÃO y(t)                                      |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[6].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[6].newLabel("Beta", 17, 15, 35, "Beta.png", True)
dados[6].newLabel("Theta", 17, 15, 70, "Theta.png", True)
dados[6].newLabel("v0", 17, 15, 105, "v0.png", True)
dados[6].newLabel("Gravidade", 17, 15, 140, "Gravidade.png", True)
dados[6].newLabel("t0", 17, 15, 175, "t0_y.png", True)
dados[6].newLabel("t1", 17, 15, 210, "t1_y.png", True)
dados[6].newLabel("Passo", 14, 13, 245, "Passo_y.png", True)

dados[6].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[6].newLabel("Unidade02", 17, 265, 70, "GrausU02.png", False)
dados[6].newLabel("Unidade03", 17, 265, 105, "VelU03.png", False)
dados[6].newLabel("Unidade04", 17, 265, 140, "GravU04.png", False)
dados[6].newLabel("Unidade05", 17, 265, 175, "TimeU05.png", False)
dados[6].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[6].newLabel("Unidade07", 17, 265, 245, "TimeU07.png", False)
dados[6].newLabel("MetodoAna", 17, 110, 277, "None", False)

dados[6].newButton(((dados[6].divWidth/2 + 75) - 60),
                    ((dados[6].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodNum",
                    f[43])

dados[6].newButton(((dados[6].divWidth/2 + 75) - 60),
                    ((dados[6].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[28])

dados[6].newButton(((dados[6].divWidth/2 - 75) - 60),
                    ((dados[6].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[73])

dados[6].newButton(((dados[6].divWidth/2 + 75) - 60),
                    ((dados[6].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[21])

dados[6].newButton(((dados[6].divWidth/2 - 75) - 60),
                    ((dados[6].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[14])

dados[6].newButton(((dados[6].divWidth/2 - 75) - 60),
                    ((dados[6].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[36])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                        TRAJETÓRIA                                         |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[7].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[7].newLabel("b", 17, 15, 35, "Beta.png", True)
dados[7].newLabel("c", 17, 15, 70, "Gamma.png", True)
dados[7].newLabel("angulo", 17, 15, 105, "ThetaNumerico.png", True)
dados[7].newLabel("v0", 17, 15, 140, "v0Numerico.png", True)
dados[7].newLabel("g", 17, 15, 175, "GravidadeNumerica.png", True)
dados[7].newLabel("Passo", 14, 13, 210, "PassoNumerico.png", True)
dados[7].newLabel("Limite", 7, 13, 245, "LimiteNumerico.png", True)

dados[7].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[7].newLabel("Unidade02", 17, 265, 70, "GammaU02.png", False)
dados[7].newLabel("Unidade03", 17, 265, 105, "GrausU03.png", False)
dados[7].newLabel("Unidade04", 17, 265, 140, "VelU04.png", False)
dados[7].newLabel("Unidade05", 17, 265, 175, "GravU05.png", False)
dados[7].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[7].newLabel("Unidade07", 17, 265, 245, "MetroU07.png", False)
dados[7].newLabel("MetodoNum", 17, 110, 277, "None", False)

dados[7].newButton(((dados[7].divWidth/2 + 75) - 60),
                    ((dados[7].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodAna",
                    f[44])

dados[7].newButton(((dados[7].divWidth/2 + 75) - 60),
                    ((dados[7].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[22])

dados[7].newButton(((dados[7].divWidth/2 + 75) - 60),
                    ((dados[7].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[60])

dados[7].newButton(((dados[7].divWidth/2 - 75) - 60),
                    ((dados[7].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[67])

dados[7].newButton(((dados[7].divWidth/2 - 75) - 60),
                    ((dados[7].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[51])

dados[7].newButton(((dados[7].divWidth/2 - 75) - 60),
                    ((dados[7].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[74])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                       POSIÇÃO x(t)                                        |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[8].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[8].newLabel("b", 17, 15, 35, "Beta.png", True)
dados[8].newLabel("c", 17, 15, 70, "Gamma.png", True)
dados[8].newLabel("angulo", 17, 15, 105, "ThetaNumerico.png", True)
dados[8].newLabel("v0", 17, 15, 140, "v0Numerico.png", True)
dados[8].newLabel("g", 17, 15, 175, "GravidadeNumerica.png", True)
dados[8].newLabel("Passo", 14, 13, 210, "PassoNumerico.png", True)
dados[8].newLabel("Limite", 7, 13, 245, "LimiteTempo.png", True)

dados[8].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[8].newLabel("Unidade02", 17, 265, 70, "GammaU02.png", False)
dados[8].newLabel("Unidade03", 17, 265, 105, "GrausU03.png", False)
dados[8].newLabel("Unidade04", 17, 265, 140, "VelU04.png", False)
dados[8].newLabel("Unidade05", 17, 265, 175, "GravU05.png", False)
dados[8].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[8].newLabel("Unidade07", 17, 265, 245, "TimeU07.png", False)
dados[8].newLabel("MetodoNum", 17, 110, 277, "None", False)

dados[8].newButton(((dados[8].divWidth/2 + 75) - 60),
                    ((dados[8].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodAna",
                    f[45])

dados[8].newButton(((dados[8].divWidth/2 + 75) - 60),
                    ((dados[8].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[23])

dados[8].newButton(((dados[8].divWidth/2 + 75) - 60),
                    ((dados[8].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[61])

dados[8].newButton(((dados[8].divWidth/2 - 75) - 60),
                    ((dados[8].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[75])

dados[8].newButton(((dados[8].divWidth/2 - 75) - 60),
                    ((dados[8].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[68])

dados[8].newButton(((dados[8].divWidth/2 - 75) - 60),
                    ((dados[8].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[52])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                       POSIÇÃO y(t)                                        |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[9].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[9].newLabel("b", 17, 15, 35, "Beta.png", True)
dados[9].newLabel("c", 17, 15, 70, "Gamma.png", True)
dados[9].newLabel("angulo", 17, 15, 105, "ThetaNumerico.png", True)
dados[9].newLabel("v0", 17, 15, 140, "v0Numerico.png", True)
dados[9].newLabel("g", 17, 15, 175, "GravidadeNumerica.png", True)
dados[9].newLabel("Passo", 14, 13, 210, "PassoNumerico.png", True)
dados[9].newLabel("Limite", 7, 13, 245, "LimiteTempo.png", True)

dados[9].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[9].newLabel("Unidade02", 17, 265, 70, "GammaU02.png", False)
dados[9].newLabel("Unidade03", 17, 265, 105, "GrausU03.png", False)
dados[9].newLabel("Unidade04", 17, 265, 140, "VelU04.png", False)
dados[9].newLabel("Unidade05", 17, 265, 175, "GravU05.png", False)
dados[9].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[9].newLabel("Unidade07", 17, 265, 245, "TimeU07.png", False)
dados[9].newLabel("MetodoNum", 17, 110, 277, "None", False)

dados[9].newButton(((dados[9].divWidth/2 + 75) - 60),
                    ((dados[9].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodAna",
                    f[46])

dados[9].newButton(((dados[9].divWidth/2 + 75) - 60),
                    ((dados[9].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[24])

dados[9].newButton(((dados[9].divWidth/2 - 75) - 60),
                    ((dados[9].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[69])

dados[9].newButton(((dados[9].divWidth/2 + 75) - 60),
                    ((dados[9].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[62])

dados[9].newButton(((dados[9].divWidth/2 - 75) - 60),
                    ((dados[9].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[53])


dados[9].newButton(((dados[9].divWidth/2 - 75) - 60),
                    ((dados[9].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[76])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      VELOCIDADE x(t)                                      |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[10].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[10].newLabel("b", 17, 15, 35, "Beta.png", True)
dados[10].newLabel("c", 17, 15, 70, "Gamma.png", True)
dados[10].newLabel("angulo", 17, 15, 105, "ThetaNumerico.png", True)
dados[10].newLabel("v0", 17, 15, 140, "v0Numerico.png", True)
dados[10].newLabel("g", 17, 15, 175, "GravidadeNumerica.png", True)
dados[10].newLabel("Passo", 14, 13, 210, "PassoNumerico.png", True)
dados[10].newLabel("Limite", 7, 13, 245, "LimiteTempo.png", True)

dados[10].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[10].newLabel("Unidade02", 17, 265, 70, "GammaU02.png", False)
dados[10].newLabel("Unidade03", 17, 265, 105, "GrausU03.png", False)
dados[10].newLabel("Unidade04", 17, 265, 140, "VelU04.png", False)
dados[10].newLabel("Unidade05", 17, 265, 175, "GravU05.png", False)
dados[10].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[10].newLabel("Unidade07", 17, 265, 245, "TimeU07.png", False)
dados[10].newLabel("MetodoNum", 17, 110, 277, "None", False)

dados[10].newButton(((dados[10].divWidth/2 + 75) - 60),
                    ((dados[10].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodAna",
                    f[47])

dados[10].newButton(((dados[10].divWidth/2 + 75) - 60),
                    ((dados[10].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[25])

dados[10].newButton(((dados[10].divWidth/2 + 75) - 60),
                    ((dados[10].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[63])

dados[10].newButton(((dados[10].divWidth/2 - 75) - 60),
                    ((dados[10].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[77])

dados[10].newButton(((dados[10].divWidth/2 - 75) - 60),
                    ((dados[10].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[70])

dados[10].newButton(((dados[10].divWidth/2 - 75) - 60),
                    ((dados[10].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[54])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      VELOCIDADE y(t)                                      |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[11].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[11].newLabel("b", 17, 15, 35, "Beta.png", True)
dados[11].newLabel("c", 17, 15, 70, "Gamma.png", True)
dados[11].newLabel("angulo", 17, 15, 105, "ThetaNumerico.png", True)
dados[11].newLabel("v0", 17, 15, 140, "v0Numerico.png", True)
dados[11].newLabel("g", 17, 15, 175, "GravidadeNumerica.png", True)
dados[11].newLabel("Passo", 14, 13, 210, "PassoNumerico.png", True)
dados[11].newLabel("Limite", 7, 13, 245, "LimiteTempo.png", True)

dados[11].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[11].newLabel("Unidade02", 17, 265, 70, "GammaU02.png", False)
dados[11].newLabel("Unidade03", 17, 265, 105, "GrausU03.png", False)
dados[11].newLabel("Unidade04", 17, 265, 140, "VelU04.png", False)
dados[11].newLabel("Unidade05", 17, 265, 175, "GravU05.png", False)
dados[11].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[11].newLabel("Unidade07", 17, 265, 245, "TimeU07.png", False)
dados[11].newLabel("MetodoNum", 17, 110, 277, "None", False)

dados[11].newButton(((dados[11].divWidth/2 + 75) - 60),
                    ((dados[11].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodAna",
                    f[48])

dados[11].newButton(((dados[11].divWidth/2 + 75) - 60),
                    ((dados[11].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[26])

dados[11].newButton(((dados[11].divWidth/2 - 75) - 60),
                    ((dados[11].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[71])

dados[11].newButton(((dados[11].divWidth/2 + 75) - 60),
                    ((dados[11].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[64])

dados[11].newButton(((dados[11].divWidth/2 - 75) - 60),
                    ((dados[11].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[55])

dados[11].newButton(((dados[11].divWidth/2 - 75) - 60),
                    ((dados[11].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[78])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      ACELERAÇÃO x(t)                                      |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[12].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[12].newLabel("b", 17, 15, 35, "Beta.png", True)
dados[12].newLabel("c", 17, 15, 70, "Gamma.png", True)
dados[12].newLabel("angulo", 17, 15, 105, "ThetaNumerico.png", True)
dados[12].newLabel("v0", 17, 15, 140, "v0Numerico.png", True)
dados[12].newLabel("g", 17, 15, 175, "GravidadeNumerica.png", True)
dados[12].newLabel("Passo", 14, 13, 210, "PassoNumerico.png", True)
dados[12].newLabel("Limite", 7, 13, 245, "LimiteTempo.png", True)

dados[12].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[12].newLabel("Unidade02", 17, 265, 70, "GammaU02.png", False)
dados[12].newLabel("Unidade03", 17, 265, 105, "GrausU03.png", False)
dados[12].newLabel("Unidade04", 17, 265, 140, "VelU04.png", False)
dados[12].newLabel("Unidade05", 17, 265, 175, "GravU05.png", False)
dados[12].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[12].newLabel("Unidade07", 17, 265, 245, "TimeU07.png", False)
dados[12].newLabel("MetodoNum", 17, 110, 277, "None", False)

dados[12].newButton(((dados[12].divWidth/2 + 75) - 60),
                    ((dados[12].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodAna",
                    f[49])

dados[12].newButton(((dados[12].divWidth/2 + 75) - 60),
                    ((dados[12].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[27])

dados[12].newButton(((dados[12].divWidth/2 - 75) - 60),
                    ((dados[12].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[72])

dados[12].newButton(((dados[12].divWidth/2 + 75) - 60),
                    ((dados[12].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[65])

dados[12].newButton(((dados[12].divWidth/2 - 75) - 60),
                    ((dados[12].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[56])

dados[12].newButton(((dados[12].divWidth/2 - 75) - 60),
                    ((dados[12].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[79])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                      ACELERAÇÃO y(t)                                      |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

dados[13].newLabel("Nome", 24, 15, 5, "Nome.png", True)
dados[13].newLabel("b", 17, 15, 35, "Beta.png", True)
dados[13].newLabel("c", 17, 15, 70, "Gamma.png", True)
dados[13].newLabel("angulo", 17, 15, 105, "ThetaNumerico.png", True)
dados[13].newLabel("v0", 17, 15, 140, "v0Numerico.png", True)
dados[13].newLabel("g", 17, 15, 175, "GravidadeNumerica.png", True)
dados[13].newLabel("Passo", 14, 13, 210, "PassoNumerico.png", True)
dados[13].newLabel("Limite", 7, 13, 245, "LimiteTempo.png", True)

dados[13].newLabel("Unidade01", 17, 265, 35, "BetaU01.png", False)
dados[13].newLabel("Unidade02", 17, 265, 70, "GammaU02.png", False)
dados[13].newLabel("Unidade03", 17, 265, 105, "GrausU03.png", False)
dados[13].newLabel("Unidade04", 17, 265, 140, "VelU04.png", False)
dados[13].newLabel("Unidade05", 17, 265, 175, "GravU05.png", False)
dados[13].newLabel("Unidade06", 17, 265, 210, "TimeU06.png", False)
dados[13].newLabel("Unidade07", 17, 265, 245, "TimeU07.png", False)
dados[13].newLabel("MetodoNum", 17, 110, 277, "None", False)

dados[13].newButton(((dados[13].divWidth/2 + 75) - 60),
                    ((dados[13].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "MethodAna",
                    f[50])

dados[13].newButton(((dados[13].divWidth/2 + 75) - 60),
                    ((dados[13].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Limp",
                    f[28])

dados[13].newButton(((dados[13].divWidth/2 - 75) - 60),
                    ((dados[13].divHeight/2 + 145) - 25),
                    120,
                    50,
                    "Import",
                    f[73])

dados[13].newButton(((dados[13].divWidth/2 + 75) - 60),
                    ((dados[13].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Remove",
                    f[66])

dados[13].newButton(((dados[13].divWidth/2 - 75) - 60),
                    ((dados[13].divHeight/2 + 205) - 25),
                    120,
                    50,
                    "Calc",
                    f[57])

dados[13].newButton(((dados[13].divWidth/2 - 75) - 60),
                    ((dados[13].divHeight/2 + 85) - 25),
                    120,
                    50,
                    "Export",
                    f[80])

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                   CALCULADORA PARA BETA                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

calculadoraBeta.newLabel("n", 17, 15, 15, "Eta.png", True)
calculadoraBeta.newLabel("r", 17, 15, 50, "rBeta.png", True)
calculadoraBeta.newLabel("m", 17, 15, 85, "mBeta.png", True)

calculadoraBeta.newButton(((calculadoraBeta.divWidth/2) - 45),
                           ((calculadoraBeta.divHeight/2 + 55) - 18.75),
                           90,
                           37.5,
                           "SmallCalc",
                           f[58])

calculadoraBeta.newLabel("Resultado", 17, 15, 201.5, "BetaCalculator.png", True)

calculadoraBeta.newLabel("Unit01", 17, 265, 15, "etaUnit.png", False)
calculadoraBeta.newLabel("Unit02", 17, 265, 50, "rBetaUnit.png", False)
calculadoraBeta.newLabel("Unit03", 17, 265, 85, "mBetaUnit.png", False)
calculadoraBeta.newLabel("Unit04", 17, 265, 201.5, "betaCalUnit.png", False)

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                  CALCULADORA PARA GAMMA                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

calculadoraGamma.newLabel("C", 21, 15, 15, "C.png", True)
calculadoraGamma.newLabel("p", 16, 15, 50, "Rho.png", True)
calculadoraGamma.newLabel("r", 17, 15, 85, "rGamma.png", True)
calculadoraGamma.newLabel("m", 17, 15, 120, "mGamma.png", True)

calculadoraGamma.newButton(((calculadoraGamma.divWidth/2) - 45),
                           ((calculadoraGamma.divHeight/2 + 55) - 18.75),
                           90,
                           37.5,
                           "SmallCalc",
                           f[59])

calculadoraGamma.newLabel("Resultado", 17, 15, 201.5, "GammaCalculator.png", True)

calculadoraGamma.newLabel("Unit01", 17, 255, 50, "rhoUnit.png", False)
calculadoraGamma.newLabel("Unit02", 17, 265, 85, "rGammaUnit.png", False)
calculadoraGamma.newLabel("Unit03", 17, 265, 120, "mGammaUnit.png", False)
calculadoraGamma.newLabel("Unit04", 17, 265, 201.5, "gammaCalUnit.png", False)

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                   GRÁFICOS DAS DIVISÕES                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================
#
#     Aqui está sendo criado todos os gráficos de todos as divisões de dados.
# Dessa forma é possível ver uma representação gráfica da utilização das equa-
# ções.
#     A função createGraph() recebe em seus parâmetros: O nome do gráfico, no-
# me

graficos[0].createGraph("TRAJETÓRIA DO MOVIMENTO PARABÓLICO COM RESISTÊNCIA DO AR", "Posição X (m)", "Posição Y (m)")
graficos[1].createGraph("x(t) DO MOVIMENTO PARABÓLICO COM RESISTÊNCIA DO AR", "Tempo (s)", "Posição X (m)")
graficos[2].createGraph("y(t) DO MOVIMENTO PARABÓLICO COM RESISTÊNCIA DO AR", "Tempo (s)", "Posição Y (m)")
graficos[3].createGraph("Vx(t) DO MOVIMENTO PARABÓLICO COM RESISTÊNCIA DO AR", "Tempo (s)", "Velocidade X (m/s)")
graficos[4].createGraph("Vy(t) DO MOVIMENTO PARABÓLICO COM RESISTÊNCIA DO AR", "Tempo (s)", "Velocidade Y (m/s)")
graficos[5].createGraph("Ax(t) DO MOVIMENTO PARABÓLICO COM RESISTÊNCIA DO AR", "Tempo (s)", "Aceleração X (m/s²)")
graficos[6].createGraph("Ay(t) DO MOVIMENTO PARABÓLICO COM RESISTÊNCIA DO AR", "Tempo (s)", "Aceleração Y (m/s²)")

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                    ATUALIZA AS DIVISÕES                                   |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

menu.updateDivision()
calculadoraBeta.updateDivision()
calculadoraGamma.updateDivision()
dados[0].updateDivision()
dados[1].updateDivision()
dados[2].updateDivision()
dados[3].updateDivision()
dados[4].updateDivision()
dados[5].updateDivision()
dados[6].updateDivision()
dados[7].updateDivision()
dados[8].updateDivision()
dados[9].updateDivision()
dados[10].updateDivision()
dados[11].updateDivision()
dados[12].updateDivision()
dados[13].updateDivision()
helpDiv.updateDivision()

# =============================================================================================
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# |                                       BIND DOS BOTÕES                                     |
# |                                                                                           |
# |                                                                                           |
# |                                                                                           |
# =============================================================================================

menu.background.bind('<Motion>', menu.isOver)
menu.background.bind('<Button-1>', menu.isClicked)
dados[0].background.bind('<Motion>', dados[0].isOver)
dados[0].background.bind('<Button-1>', dados[0].isClicked)
dados[1].background.bind('<Motion>', dados[1].isOver)
dados[1].background.bind('<Button-1>', dados[1].isClicked)
dados[2].background.bind('<Motion>', dados[2].isOver)
dados[2].background.bind('<Button-1>', dados[2].isClicked)
dados[3].background.bind('<Motion>', dados[3].isOver)
dados[3].background.bind('<Button-1>', dados[3].isClicked)
dados[4].background.bind('<Motion>', dados[4].isOver)
dados[4].background.bind('<Button-1>', dados[4].isClicked)
dados[5].background.bind('<Motion>', dados[5].isOver)
dados[5].background.bind('<Button-1>', dados[5].isClicked)
dados[6].background.bind('<Motion>', dados[6].isOver)
dados[6].background.bind('<Button-1>', dados[6].isClicked)
dados[7].background.bind('<Motion>', dados[7].isOver)
dados[7].background.bind('<Button-1>', dados[7].isClicked)
dados[8].background.bind('<Motion>', dados[8].isOver)
dados[8].background.bind('<Button-1>', dados[8].isClicked)
dados[9].background.bind('<Motion>', dados[9].isOver)
dados[9].background.bind('<Button-1>', dados[9].isClicked)
dados[10].background.bind('<Motion>', dados[10].isOver)
dados[10].background.bind('<Button-1>', dados[10].isClicked)
dados[11].background.bind('<Motion>', dados[11].isOver)
dados[11].background.bind('<Button-1>', dados[11].isClicked)
dados[12].background.bind('<Motion>', dados[12].isOver)
dados[12].background.bind('<Button-1>', dados[12].isClicked)
dados[13].background.bind('<Motion>', dados[13].isOver)
dados[13].background.bind('<Button-1>', dados[13].isClicked)
calculadoraBeta.background.bind('<Motion>', calculadoraBeta.isOver)
calculadoraBeta.background.bind('<Button-1>', calculadoraBeta.isClicked)
calculadoraGamma.background.bind('<Motion>', calculadoraGamma.isOver)
calculadoraGamma.background.bind('<Button-1>', calculadoraGamma.isClicked)
helpDiv.background.bind('<Motion>', helpDiv.isOver)
helpDiv.background.bind('<Button-1>', helpDiv.isClicked)

root.mainloop()
```
